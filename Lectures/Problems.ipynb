```python
palindrome = 0

for i in range(100,1000):
    for j in range(100,1000):
        p = i*j
        
        if(str(p)==str(p)[::-1]):
            if p > palindrome:
                palindrome = p
                

print(f"{palindrome} is the largest palindrome number")
```

```python
count = 1
num = None
n = 3

while True:
    flag = True
    for i in range(2,n//2):
        if(n%i==0):
            flag = False
            break
    if(flag):
        count+=1
    if(count==10001):
        num = n
        break
    n+=1
    
print(num)
```

```python
square_sum = 0
sum_square = 0
for i in range(1,101):
    
    square_sum+=i**2
    
    sum_square+=i
    
print(sum_square**2-square_sum)
```

```python
fact = 1
total = 0 

for i in range(1,101):
    fact*=i
    
while fact>0:
    n = fact%10
    
    total+=n
    
    fact = fact//10

print(total)
```
