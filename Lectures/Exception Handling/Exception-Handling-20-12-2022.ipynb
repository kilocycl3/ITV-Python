# Exception Handling


**try: Executable code**

**except: Handle the error occured in try block**

**else: Executable code if exception doesnt occurs**

**finally: This blocks always executes**

```python
try:
    print(s)
except:
    print("Variable not defined")
```

```python
try:
    print(name)
except Exception as e:
    print(e)
```

```python
try:
    l = [1,2,3]
#     print(l[4])
    print(s)
except NameError as e1:
    print(e1)
except IndexError as e2:
    print(e2)
```

```python
try:
    l = [1,2,3]
    print(l[4])
    print(s)
except NameError as e1:
    print(e1)
except IndexError as e2:
    print(e2)
```

```python
try:
    with open('data.txt') as f:
        print(f.read())
        
except FileNotFoundError as e1:
    print("File Error:",e1)
```

```python
import os

try:
    os.mkdir("Directory")
    
except FileExistsError as fe:
    print("Folder Error:",fe)
```

```python
try:
    n1 = 100
    n2 = 0
    print(n1/n2)
    
except ZeroDivisionError as ze:
    print("Number is not divisible by zero")
    print(ze)
```

```python
try:
    n1 = 100
    n2 = 0
    print(n1/n2)
    
except Exception as ze:
    raise Exception("Number not divisible by zero")
    
```
